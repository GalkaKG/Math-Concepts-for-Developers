## Probability and combinatorics

### Live Demos

```python
results = np.random.randint(1, 7, size=100)
```

```python
plt.plot(results)
plt.xlabel("# trial")
plt.ylabel("Result")
plt.show()
```

```python
results2 = np.random.randint(1, 7, size=1000)
```

```python
plt.figure(figsize = (40, 5))
plt.plot(results2)
plt.xlabel("# trial")
plt.ylabel("Result")
plt.show()
```

```python
results2
```

```python
counts = dict()
for result in results2:
    if not result in counts:
        counts[result] = 1
    else:
        counts[result] += 1
```

```python
counts
```

```python
counts2 = Counter(results2)
```

```python
counts2
```

```python
plt.hist(results2, bins = range(1, 8))  # 
```

```python
[v / len(results2) for v in counts.values()]  #    
#  ,      
```

```python
plt.bar(counts.keys(), counts.values())  # bar chart (x, y)

plt.xlabel("Value")
plt.ylabel("Frequency")
```
